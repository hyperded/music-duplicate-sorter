```python
def has_cover_art(info):
    streams = info.get("streams", [])

    for s in streams:
        if s.get("codec_type") == "video":
            if s.get("disposition", {}).get("attached_pic") == 1:
                return True
    return False

def has_lyrics(audio):
    tags = audio.get("format", {}).get("tags", {})
    if any(k.lower().startswith("lyrics") for k in tags):
        return True
    return False
```

```python
def logic_handler(music_dir):
    current_highest_bit_rate = 0
    keep = {}
    for file in music_dir:
        criteria = 0
        audio = ffmpeg.probe(str(file))
        bit_rate = int(audio['streams'][0]['bit_rate'])
        current_highest_bit_rate = bit_rate if (current_highest_bit_rate < bit_rate) else current_highest_bit_rate
        if has_lyrics(audio):
            criteria += 2
        if bit_rate == current_highest_bit_rate:
            criteria += 1
        if has_cover_art(audio):
            criteria += 1
        keep[file] = criteria
    file_dir_to_keep, __ = max(keep.items(), key=lambda x: (x[1], -len(str(x[0]))))
    return file_dir_to_keep
        
```

```python

audio_extensions = {".mp3", ".flac", ".ogg", ".wav", ".m4a", ".aac"}
FILE_NAMES = [
    f for f in os.listdir(r"C:\Users\ASUS\Music")
    if os.path.splitext(f)[1].lower() in audio_extensions
]

FILE_NAMES             
```

```python
import pandas as pd
#duration, genre
input_dir = r"C:\Users\ASUS\Music"
df = pd.DataFrame(columns=['artist', 'title', 'album', 'date', 'duration', 'genre', 'file_name' ])
rows = [] 
for name in FILE_NAMES:
    file_path = os.path.join(input_dir, name)
    try:
        audio = ffmpeg.probe(file_path)
    except Exception as e:
        print(f"Skipped {name}: {e}")
        continue

    tags = audio.get('format', {}).get('tags', {})
    fmt = audio.get("format", {})

    duration = audio.get('format', {}).get('duration', None)    
    rows.append({
        "artist": tags.get("artist"),
        "title": tags.get("title"),
        "album": tags.get("album"),
        "date": tags.get("date"),
        "duration": fmt.get("duration"),
        "genre": tags.get("genre"),
        "file_name": name
    })
df = pd.DataFrame(rows)
```

```python
from rapidfuzz import fuzz
g = df.groupby('artist')['file_name'].apply(list).reset_index()

sec_map = {}
for _, row in g.iterrows():
    matched = False
    songs = row['file_name']
    
    if len(songs) > 1:
        for i in range(len(songs)):
            for j in range(i+1, len(songs)):
                s1, s2 = Path(songs[i]).stem, Path(songs[j]).stem
                score = fuzz.partial_ratio(s1, s2)
                if score > 95:
                    sec_map[s1] = s2

# dupe_df = pd.DataFrame(results)                
```
